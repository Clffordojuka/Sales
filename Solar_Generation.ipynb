```python
# import libraries

import mysql.connector
from mysql.connector import Error
import pandas as pd
```

```python
#create server connection
def create_server_connection(host_name, user_name, user_password):
    connection = None
    try:
        connection = mysql.connector.connect(
            host = host_name,
            user = user_name,
            passwd = user_password
        )
        print(" MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")
    return connection

#Use SQL Terminal Password

pw = ""

#Database name
db = "Solar_Generation"
connection = create_server_connection("localhost", "root", pw)
```

```python
#create database Solar_Generation
def create_database(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Database created successfully")
    except Error as err:
        print(f"Error: '{err}'")
create_database_query = "create database Solar_Generation"
create_database(connection, create_database_query)
```

```python
# connect to database

def create_db_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
           host = host_name,
           user = user_name,
           passwd = user_password,
           database = db_name)
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")
    return connection
```

```python
# Execute the sql queries

def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query was successful")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
create_SolarElectricity_table = """
    create table SolarElectricity(
    Hour INT,  -- Represents the hour of the day (0-23)
    DateHourStart DATETIME,  -- Represents the date and hour start, format 'yyyy-mm-dd hh:mm'
    SolarElectricityGeneration DECIMAL(10,7),  -- Represents solar electricity generation in kWh
    ElectricityUsage DECIMAL(10,7)  -- Represents electricity usage in kWh
    );
"""
#connect to the database
connection = create_db_connection("localhost", "root", pw, db)
execute_query(connection, create_SolarElectricity_table)
```

```python
# reading queries
def read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as err:
        print(f"Error: '{err}'")
```

```python
# Using the select statement
Q1 = """
select * from SolarElectricity;
"""
connection = create_db_connection("localhost", "root", pw, db)
results = read_query(connection, Q1)
for result in results:
    print(result)
```

```python
from_db = []

for result in results:
    result = list(result)
    from_db.append(result)
columns = ["Hour",  "DateHourStart", "SolarElectricityGeneration", "ElectricityUsage"]
df = pd.DataFrame(from_db, columns = columns)

df['SolarElectricityGeneration'] = pd.to_numeric(df['SolarElectricityGeneration'], errors='coerce')
df['ElectricityUsage'] = pd.to_numeric(df['ElectricityUsage'], errors='coerce')

display(df)
```

```python
import numpy as np

# Calculate the electricity needed to be bought
electricity_needed = np.maximum(df['ElectricityUsage'] - df['SolarElectricityGeneration'], 0)

df['ElectricityNeeded'] = electricity_needed

# Display the updated DataFrame
df.head()
```

```python
# Calculate the excess solar generation
excess_solar_generation = np.maximum(df['SolarElectricityGeneration'] - df['ElectricityUsage'], 0)

df['ExcessSolarGeneration'] = excess_solar_generation

# Display the updated DataFrame
df.head()
```

```python
# Display the first few rows to inspect the new column
df[['Hour', 'DateHourStart', 'SolarElectricityGeneration', 'ElectricityUsage', 'ElectricityNeeded', 'ExcessSolarGeneration']].head(20)
```

```python
# Calculate the average excess solar generation for each hour
avg_excess_by_hour = df.groupby('Hour')['ExcessSolarGeneration'].mean()

# Plot the average excess solar generation by hour
plt.figure(figsize=(10, 6))
avg_excess_by_hour.plot(kind='bar', color='orange')
plt.title('Average Excess Solar Generation by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Excess Solar Generation (kWh)')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
# Calculate the total excess solar generation
total_excess_solar = df['ExcessSolarGeneration'].sum()

print(f"Total Excess Solar Generation: {total_excess_solar:.2f} kWh")
```

```python
# Find the hour with the maximum average excess solar generation
max_excess_hour = avg_excess_by_hour.idxmax()
max_excess_value = avg_excess_by_hour.max()

print(f"Hour with Maximum Average Excess Solar Generation: {max_excess_hour}:00")
print(f"Maximum Average Excess Solar Generation: {max_excess_value:.2f} kWh")
```

```python
# Maximum battery capacity in kWh
max_battery_capacity = 12.5

# Initialize battery charge level
battery_charge = 0

# Add a new column for battery charge level
df['BatteryChargeLevel'] = 0.0
```

```python
# Loop through each row to update battery charge level
for i in range(len(df)):
    # Add excess solar generation to the battery charge
    battery_charge += df.at[i, 'ExcessSolarGeneration']
    
    # Ensure battery charge does not exceed maximum capacity
    if battery_charge > max_battery_capacity:
        battery_charge = max_battery_capacity
    
    # Update the DataFrame with the current battery charge level
    df.at[i, 'BatteryChargeLevel'] = battery_charge
```

```python
# Plot the battery charge level over time
plt.figure(figsize=(12, 6))
plt.plot(df['DateHourStart'], df['BatteryChargeLevel'], color='green')
plt.title('Cumulative Battery Charge Level Over Time')
plt.xlabel('Time')
plt.ylabel('Battery Charge Level (kWh)')
plt.grid(True)
plt.show()
```

```python
# Step 1: Calculate initial electricity bought from the provider
df['ElectricityNeeded'] = np.maximum(df['ElectricityUsage'] - df['SolarElectricityGeneration'], 0)

# Step 2: Calculate electricity bought when using the battery
df['ElectricityBoughtWithBattery'] = df['ElectricityNeeded'] - df['ExcessSolarGeneration']

# Step 3: Ensure that the electricity bought does not go below zero
df['ElectricityBoughtWithBattery'] = df['ElectricityBoughtWithBattery'].apply(lambda x: max(x, 0))

# Display the updated DataFrame with relevant columns
display(df[['DateHourStart', 'ElectricityUsage', 'SolarElectricityGeneration', 'ElectricityNeeded', 'ExcessSolarGeneration', 'ElectricityBoughtWithBattery']])
```

```python
# Step 1: Define the electricity price per kWh
electricity_price_per_kwh = 0.17

# Step 2: Calculate the cost of electricity bought without a battery
df['CostWithoutBattery'] = df['ElectricityNeeded'] * electricity_price_per_kwh

# Step 3: Calculate the cost of electricity bought with a battery
df['CostWithBattery'] = df['ElectricityBoughtWithBattery'] * electricity_price_per_kwh

# Step 4: Calculate the potential savings from installing a battery
df['Savings'] = df['CostWithoutBattery'] - df['CostWithBattery']

# Display the relevant columns to verify the results
display(df[['DateHourStart', 'ElectricityNeeded', 'ElectricityBoughtWithBattery', 'CostWithoutBattery', 'CostWithBattery', 'Savings']])
```

```python

# Step 1: Convert DateHourStart to DateTime and extract the month
df['Month'] = pd.to_datetime(df['DateHourStart']).dt.to_period('M')

# Step 2: Aggregate data by month
monthly_aggregated = df.groupby('Month').agg({
    'SolarElectricityGeneration': 'sum',
    'ElectricityUsage': 'sum',
    'ElectricityNeeded': 'sum',
    'ElectricityBoughtWithBattery': 'sum'
}).reset_index()

# Step 3: Visualize the aggregated data using a bar plot
plt.figure(figsize=(10, 6))

# Plot total solar generation
plt.bar(monthly_aggregated['Month'].astype(str), monthly_aggregated['SolarElectricityGeneration'], color='goldenrod', label='Total Solar Generation')

# Plot total electricity usage
plt.bar(monthly_aggregated['Month'].astype(str), monthly_aggregated['ElectricityUsage'], color='skyblue', label='Total Electricity Usage')

# Plot electricity bought without a battery
plt.bar(monthly_aggregated['Month'].astype(str), monthly_aggregated['ElectricityNeeded'], color='red', label='Electricity Bought Without Battery')

# Plot electricity bought with a battery
plt.bar(monthly_aggregated['Month'].astype(str), monthly_aggregated['ElectricityBoughtWithBattery'], color='green', label='Electricity Bought With Battery')

# Customize the plot
plt.xlabel('Month')
plt.ylabel('kWh')
plt.title('Monthly Aggregated Solar Generation, Electricity Usage, and Electricity Needed')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```
